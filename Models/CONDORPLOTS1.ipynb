```python
#Método do gradiente

function Grad1(X)
    x = X[1]
    y = X[2]
    soma1 = 0
    soma2 = 0
    for i = 1:length(Pontos[:,1])
        soma1 = soma1 + (x-Pontos[i,1])/(sqrt((x-Pontos[i,1])^2 + (y-Pontos[i,2])^2))
        soma2 = soma2 + (y-Pontos[i,2])/(sqrt((x-Pontos[i,1])^2 + (y-Pontos[i,2])^2))
    end
    V = [soma1, soma2]
    return V
end
```

```python
# Método do Gradiente sem peso
x0 = [-25, -49]
xk = x0
k = 0
while norm(Grad1(xk)) > 1e-8
    dk = -Grad1(xk)
    gama = 0.8
    eta = 0.5
    t = 1
    while f(xk + t*dk) > f(xk) + eta*t*dot(-dk,dk)
        t = gama*t
    end
    xk = xk + t*dk
    k = k+ 1
end
print(xk)
print(k)
```

```python
# Hessiana

function Hessi(X)
    x = X[1]
    y = X[2]
    soma1 = 0
    soma2 = 0
    soma3 = 0
    for i = 1:length(Pontos[:,1])
        soma1 = soma1 + ((x-Pontos[i,1])^2 + (y-Pontos[i,2])^2)^(-1/2) - (((x-Pontos[i,1])^2))/((((x-Pontos[i,1])^2 + (y-Pontos[i,2])^2))^(3/2))
        soma2 = soma2 + ((x-Pontos[i,1])^2 + (y-Pontos[i,2])^2)^(-1/2) - (((y-Pontos[i,2])^2))/((((x-Pontos[i,1])^2 + (y-Pontos[i,2])^2))^(3/2))
        soma3 = soma3 - ((x-Pontos[i,1])*(y-Pontos[i,2])^2)/((((y-Pontos[i,2])^2))/((((x-Pontos[i,1])^2 + (y-Pontos[i,2])^2))^(3/2))) 
    end
    V = [soma1 soma3; soma3 soma2]
    return V
end
```

```python
# Método de Newton
x0 = [-25, -49]
xk = x0
k = 0
while norm(Grad1(xk)) > 1e-8
    dk = Hessi(xk)\(-Grad1(xk))
    gama = 0.8
    eta = 0.5
    t = 1
    while f(xk + t*dk) > f(xk) + eta*t*dot(Grad1(xk),dk)
        t = gama*t
    end
    xk = xk + t*dk
    k = k+ 1
end
print(xk)
print(k)
```
